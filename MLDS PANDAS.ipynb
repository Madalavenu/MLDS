```python
import pandas as pd

df = pd.DataFrame({
    'Name': ['John', 'Sarah', 'Mike', 'Emma', 'Chris', 'Anna'],
    'Department': ['IT', 'HR', 'IT', 'Sales', 'Sales', 'HR'],
    'Salary': [75000, 65000, 80000, 55000, 58000, 62000],
    'Years': [3, 5, 7, 2, 4, 6],
    'Remote': [True, False, True, False, True, False]
})
```

```python
df[df['Department'] == 'IT']
df[df['Salary'] > 60000]
df[df['Remote'] == True]
df[df['Years'] >= 5]
```

```python
df.loc[df['Age'] > 28].iloc[:2]

df.loc[2,'Salary']
df.iloc[2]['Salary']
```

```python
df[(df['Department']=='IT') & (df['Remote']==True)]

df[(df['Salary'].between(60000,75000)) & (df['Years']>3)]

df[(df['Department']!='HR') & (df['Salary']>60000)]
```

```python
df[(df['Department']=='IT') | (df['Remote']==True)]

df[(df['Salary']<60000) | (df['Salary']>70000)]

df[(df['Department']=='Sales') | (df['Years']<3)]
```

```python
import pandas as pd

df = pd.DataFrame({
    'Date': pd.date_range('2024-01-01', periods=10),
    'Region': ['North', 'South', 'East', 'West', 'North', 
               'South', 'East', 'West', 'North', 'South'],
    'Product': ['A', 'B', 'A', 'C', 'B', 
                'A', 'C', 'B', 'A', 'C'],
    'Quantity': [10, 15, 8, 20, 12, 
                 25, 18, 22, 14, 30],
    'Revenue': [1000, 1500, 800, 2000, 1200, 
                2500, 1800, 2200, 1400, 3000]
})
```

```python
df[((df['Region']=='North') | (df['Region']=='South')) & 
   (df['Revenue'] > 1200)]
```

```python
df[((df['Region']=='North') | (df['Region']=='South')) & (df['Revenue']>1200)]

df[(df['Product'].isin(['A','B'])) & (df['Quantity'].between(10,20))]

df[(df['Region']!='West') & (df['Revenue'] >= df['Revenue'].quantile(0.5))]
```
