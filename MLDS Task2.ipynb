```python
pwd = input("Enter password: ")

if len(pwd) < 8:
    print("Weak password (too short)")
elif not any(c.isupper() for c in pwd):
    print("Weak (no uppercase)")
elif not any(c.islower() for c in pwd):
    print("Weak (no lowercase)")
elif not any(c.isdigit() for c in pwd):
    print("Weak (no digit)")
else:
    print("Strong password!")
```

```python
def mask_email(email):
    if "@" not in email:
        return email
    username, domain = email.split("@")
    if len(username) <= 2:
        return email
    return username[0] + "*"*(len(username)-2) + username[-1] + "@" + domain

def mask_phone(phone):
    digits = ''.join(filter(str.isdigit, phone))
    return "*"*(len(digits)-4) + digits[-4:]

email = input("Enter email: ")
phone = input("Enter phone: ")

print("Masked Email:", mask_email(email))
print("Masked Phone:", mask_phone(phone))
```

```python
numbers = [1, 3, 2, 3, 4, 3, 2, 1, 3]
freq = {}

for num in numbers:
    freq[num] = freq.get(num, 0) + 1

max_count = 0
max_element = None

for key, value in freq.items():
    if value > max_count:
        max_count = value
        max_element = key

print(max_element)
```

```python
def is_anagram(s1, s2):
    if len(s1) != len(s2):
        return False

    d = {}
    for c in s1:
        d[c] = d.get(c, 0) + 1

    for c in s2:
        if c not in d:
            return False
        d[c] -= 1

    return all(v == 0 for v in d.values())

print(is_anagram("listen", "silent"))
```
